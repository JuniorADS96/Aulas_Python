#Listas: A estrutura ordenada e mutável

```python
#Exemplo lista

num = [1,2,3,4,5]
for n in num:
  if n % 2 == 0:
    print("Par")
  else:
    print("Impar")
```

#Codigo modificado para mostrar o numero antes do impar ou par

```python
#Exemplo lista
num = [1,2,3,4,5] #listas
#tamanho da minha variavel, ele ira somente nos numeros que tenho dentro da variavel.

for n in num:
#aqui o for tem o limite de ir até o "num" ou seja ele vai ir até o tamanho da lista

  if n % 2 == 0:
#[IF % 2 == 0] aqui está decidindo se o numero e impar ou par
    print(n, "- Par")
#aqui inserimos o "N" para que apareça o numero antes do resultado.
  else:
    print(n, "- Ímpar")
A = len(num)
#LEN é para mostrar quantos numeros tem na lista
#Colocando len dentro de uma variavel eu consigo imprimir a len junto a uma frase
print("o tamanho da lista é de", A)

num.append(15)
#Adcionei a função print para não ficar tão confuso e mostrar o que exatamente o num.append fez
print("Adcionando o número a lista ficamos com:")
print(num)

del num [3]
#Adcionei a função print para não ficar tão confuso e mostrar o que exatamente o del fez
print("Removendo o número da lista ficamos com:")
print(num)
```

#Funçoes

Len - o valor retornado é igual ao número de elementos da lista.
append - adciona novos elementos.
del - remove elementos da lista.

Faça um programa que tenha uma lista com 5 números positivos quaisquer e apresente a soma de todos os elementos.

```python
num = [1,2,3,4,5]

soma = 0

for h in num :
  soma = soma + h

print("O resultado da soma é ", soma)
```
