#APRENDENDO SOBRE INT/INPUT/PRINT

input() é usado para receber dados do usuário pelo teclado.
Por exemplo: quando queremos que a pessoa digite algo.

int() serve para transformar um valor em número inteiro.
Por exemplo: quando o usuário digita um número, mas ele vem como texto, usamos int() para converter.

👉 Exemplo simples:
Quero que o usuário digite a idade e o programa use como número:

```python
idade = int(input("Digite sua idade: "))
# input() mostra uma mensagem na tela e espera o usuário digitar algo
# o valor recebido é texto (string)
# int() transforma esse texto em número inteiro
# idade = armazena o valor digitado na variável

print("Sua idade é", idade, "anos.")
# print() exibe informações na tela
# dentro dos () colocamos textos e variáveis
# textos ficam entre aspas "" ou ''
# variáveis NÃO usam aspas
# usamos vírgula para separar texto e variável
```
